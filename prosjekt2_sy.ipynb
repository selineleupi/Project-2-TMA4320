## Exercise 1a

Show that $\Phi(x,t) = \frac{1}{\sqrt{4\pi Dt}} \cdot \int_{-\infty}^{\infty} h(y) \cdot e^{-\frac{(x-y)^2}{4Dt}} \,dy$ is a solution of the diffusion equation, $\frac{\partial\Phi(x,t)}{\partial t} = D \cdot \frac{\partial^2 \Phi(x,t)}{\partial^2 x}$.

Showing it by calculating each side separately, and seeing that the equality still holds.

<br><br>

Left side:

\begin{equation}
\frac{\partial\Phi(x,t)}{\partial t} 
= \frac{1}{\sqrt{4\pi D}} 
\cdot \left( \int_{-\infty}^{\infty} h(y) \cdot e^{-\frac{(x-y)^2}{4Dt}} \,dy \cdot \frac{\partial}{\partial t} (t^{-\frac{1}{2}})
+ \frac{1}{\sqrt{t}} \cdot \int_{-\infty}^{\infty} h(y) \cdot \frac{\partial}{\partial t} e^{-\frac{(x-y)^2}{4Dt}} \,dy \right)
\end{equation}

\begin{equation}
= \frac{1}{\sqrt{4\pi D}} 
\cdot \left( -\frac{1}{2} t^{-\frac{3}{2}} 
\cdot \int_{-\infty}^{\infty} h(y) \cdot e^{-\frac{(x-y)^2}{4Dt}}  \,dy
+ \frac{1}{\sqrt{t}} 
\cdot \int_{-\infty}^{\infty} h(y) \cdot \frac{(x-y)^2}{4Dt^2} \cdot e^{-\frac{(x-y)^2}{4Dt}} \,dy \right)
\end{equation}

\begin{equation}
= \frac{1}{\sqrt{4\pi D}t^{\frac{5}{2}}} 
\cdot \int_{-\infty}^{\infty} h(y) \cdot e^{-\frac{(x-y)^2}{4Dt}} \cdot \left( -\frac{t}{2} + \frac{(x-y)^2}{4D} \right) \,dy
\end{equation}

<br><br>
Right side:

\begin{equation}
\frac{\partial\Phi(x,t)}{\partial x} 
= \frac{1}{\sqrt{4\pi Dt}} 
\cdot \left( \int_{-\infty}^{\infty} h(y) \cdot \frac{\partial}{\partial x} \cdot e^{-\frac{(x-y)^2}{4Dt}} \,dy \right)
\end{equation}

\begin{equation}
= \frac{1}{\sqrt{4\pi Dt}} 
\cdot \left( \int_{-\infty}^{\infty} h(y) \cdot \frac{-2(x-y)}{4Dt} \cdot e^{-\frac{(x-y)^2}{4Dt}} \,dy \right)
\end{equation}

<br><br>

\begin{equation}
D \cdot \frac{\partial^2 \Phi(x,t)}{\partial^2 x} 
= \frac{D}{\sqrt{4\pi Dt}} 
\cdot \left( \int_{-\infty}^{\infty} h(y) \cdot \frac{\partial}{\partial x} \left( \frac{(y-x)}{2Dt} \cdot e^{-\frac{(x-y)^2}{4Dt}} \right) \,dy \right)
\end{equation}

\begin{equation}
= \frac{1}{\sqrt{4\pi Dt} \cdot 2t} 
\cdot \left( \int_{-\infty}^{\infty} h(y) \cdot \left( -1 + \frac{2(x-y)^2}{4Dt} \right) \cdot e^{-\frac{(x-y)^2}{4Dt}}  \,dy \right)
\end{equation}

\begin{equation}
= \frac{1}{\sqrt{4\pi D} \cdot t^\frac{5}{2}} 
\cdot \left( \int_{-\infty}^{\infty} h(y) \cdot \left( -\frac{t}{2} + \frac{(x-y)^2}{4Dt} \right) \cdot e^{-\frac{(x-y)^2}{4Dt}}  \,dy \right)
\end{equation}

\begin{equation}
= \frac{\partial\Phi(x,t)}{\partial t} \square
\end{equation}

<br><br>

Letting $\Phi(x,t)$ represent a particle distribution along the x-axis at times $t>0$, we get that the physical interpretation of $h(y)$ is a start distribution at time $t=0$:

First we noticed that $\Phi(x,t)$ is the convolution between $h(y)$ and a Gaussian function of t,

\begin{equation}
\Phi(x,t) = \frac{1}{\sqrt{4\pi Dt}} 
\cdot \int_{-\infty}^{\infty} h(y) \cdot e^{-\frac{(x-y)^2}{4Dt}} \,dy = 
\left( h *  \left( \frac{1}{\sqrt{4\pi Dt}} \cdot e^{-\frac{x^2}{4Dt}} \right)\right) (x)
\end{equation}

If we find the expected value and the variance of $\frac{1}{\sqrt{4\pi Dt}} \cdot e^{-\frac{x^2}{4Dt}}$ alone, we get the expected value to be $x=0$ and a time dependent variance equal to $2Dt$. This means that our particle distribution is spreading out in time. 

We then followed the hint and found out what happens when vi let $t\rightarrow0$:

- Using L'Hôpital's rule and letting $t\to 0$, we are sent into a continous loop of expressions with zeros in the denominator. This did not lead us any closer to an answer.
- Since variance depends on $t$, the Gussian part of the convolution approaches the dirac delta function as $t$ approaches zero. For the dirac delta function, $\delta (x-y)$, we have that its value is zero for all x, except at x=y, where it approaches infinity. We also have the following relation:

\begin{equation}
\int_{-\infty}^{\infty} \delta (y-x) \,dy = 1
\end{equation}

<br>
which leads to, 

\begin{equation}
\Phi(x,t) =  \int_{-\infty}^{\infty} h(y) \cdot \delta (y - x) \,dy = h(x) \implies \Phi(x,0) = h(x)
\end{equation}

- We also recognize that if we look at the Gaussian expression as a normal distribution for y, with E(y) = x, then the integral above is the expectation value of h(y). 
- Using Taylor expansion we get

\begin{equation}
\Phi(x,t) =  <h(y)> \approx h\left(<y> \right) + \frac{h''\left(<y> \right)}{2} \cdot Var(y)
= h\left(x \right) + \frac{h''\left(x \right)}{2} \cdot 2Dt = h\left(x\right) + h''\left(x \right) \cdot Dt
\end{equation}

We observe that $h(x)$ is a startdistribution, and that $\Phi(x,t)$ describes $h(x)$ spreading/smoothing out over time. This fits well with the fact that $\Phi(x,t)$ is a solution to the diffusion equation.

## Exercise 1b

The probability for finding a particle at position $x$ in a potential landscape, $V(x)$, is given by $P(x) = \frac{e^{-\beta V(x)}}{Z}$, where $\beta=\frac{1}{k_BT}$. Here $k_BT$ is Boltzmann's constant, and T is temperature. $Z$ is the canonical partition function of our system.

If a particle is at position $x_0$, after the next timestep, the probabilities for moving forward to $x_0+1$, staying put at $x_0$ or moving backwards to $x_0-1$, are $p^+$, $p^0$ and $p^-$ respectively.

Assuming that the proportionality constants are equal for all options, we can use $P(x)$ to find equations for each probability. The probability for a desired option is the probability for that option, divided by the sum of the probabilities for all options:

\begin{equation}
p^+ = \frac{p^+}{p^+ + p^0 + p^-} = \frac{P(x_0 + 1)}{P(x_0 + 1) + P(x_0) + P(x_0 - 1)} 
= \frac {\frac{e^{-\beta V(x_0 + 1)}}{Z}}  {\frac{e^{-\beta V(x_0 +1 )}}{Z} + \frac{e^{-\beta V(x_0)}}{Z} + \frac{e^{-\beta V(x_0)-1}}{Z}} 
= \frac{e^{-\beta V(x_0 + 1)}} {e^{-\beta V(x_0 + 1)} + e^{-\beta V(x_0)} + e^{-\beta V(x_0 - 1)}}
\end{equation}

Deviding by $e^{-\beta V(x_0 + 1)}$ in both the numerator and denominator and using that $\frac{e^x}{e^y} = e^{x-y}, we get:

\begin{equation}
p^+ = \frac{1}{1 + e^{-\beta [ V(x_0 - 1) - V(x_0 + 1) ]} + e^{-\beta [ V(x_0) - V(x_0 + 1) ]}}
\end{equation}

Deriving equations for $p^0$ and $p^-$ using the same procedure, we obtain:

\begin{equation}
p^0 = \frac{1}{1 + e^{-\beta [V(x_0 + 1) - V(x_0)]} + e^{-\beta [ V(x_0 - 1) - V(x_0) ]}}
\end{equation}

\begin{equation}
p^- = \frac{1}{1 + e^{-\beta [V(x_0 + 1) - V(x_0 - 1)]} + e^{-\beta [ V(x_0) - V(x_0 - 1) ]}}
\end{equation}

## Exercise 1c

In the case where:

- $k_BT >> |V(x+1) - V(x)|$ for any x,

We get that $\beta = \frac{1}{k_BT} << |V(x+1) - V(x)|$ for any x, and the exponent decreases to a very small number. $\lim_{x\to0}e^x = \lim_{x\to0}e^{-x} = 1$

\begin{equation}
\lim_{\beta\to 0} p^+ = \lim_{k_BT\to\infty} \frac{1}{1 + e^{-\beta [ V(x_0 - 1) - V(x_0 + 1) ]} + e^{-\beta [ V(x_0) - V(x_0 + 1) ]}} = \frac{1}{1+1+1} = \frac{1}{3}
\end{equation}

It then follows that
\begin{equation}
\lim_{\beta\to 0} p^0 = \frac{1}{3}
\end{equation}

\begin{equation}
\lim_{\beta\to 0} p^- = \frac{1}{3}
\end{equation}

For high temperatures, when particles have high internal energies relative to the potential they live in, we see that they stop obeying the potential and the probabilities become the same in all directions.


In the case where:
- $k_BT << |V(x+1) - V(x)|$ for any x

We get two different solutions depending on which direction the potential is increasing.

If $V(x+1) > V(x)$ and $\beta = \frac{1}{k_BT}$ we get:

\begin{equation}
\lim_{\beta\to\infty} p^+ = \lim_{\beta\to\infty} \frac{1} {1 + e^{-\beta [ V(x_0 - 1) - V(x_0 + 1) ]} + e^{-\beta [ V(x_0) - V(x_0 + 1) ]}} 
= \frac{1}{1 + \infty + \infty} = 0
\end{equation}

\begin{equation}
\lim_{\beta\to\infty} p^0 = \lim_{\beta\to\infty} \frac{1}{1 + e^{-\beta [V(x_0 + 1) - V(x_0)]} + e^{-\beta [ V(x_0 - 1) - V(x_0) ]}} 
= \frac{1}{1 + 0 + \infty} = 0
\end{equation}

\begin{equation}
\lim_{\beta\to\infty} p^- = \lim_{\beta\to\infty} \frac{1}{1 + e^{-\beta [V(x_0 + 1) - V(x_0 - 1)]} + e^{-\beta [ V(x_0) - V(x_0 - 1) ]}}
= \frac{1}{1 + 0 + 0} = 1
\end{equation}

Here, all particles are pushed backwards by a dominating potential.

<br>

If $V(x+1) < V(x)$ and $\beta = \frac{1}{k_BT}$ we get:

\begin{equation}
\lim_{\beta\to\infty} p^+ = \lim_{\beta\to\infty} \frac{1} {1 + e^{-\beta [ V(x_0 - 1) - V(x_0 + 1) ]} + e^{-\beta [ V(x_0) - V(x_0 + 1) ]}} 
= \frac{1}{1 + 0 + 0} = 1
\end{equation}

\begin{equation}
\lim_{\beta\to\infty} p^0 = \lim_{\beta\to\infty} \frac{1}{1 + e^{-\beta [V(x_0 + 1) - V(x_0)]} + e^{-\beta [ V(x_0 - 1) - V(x_0) ]}} 
= \frac{1}{1 + \infty + 0} = 0
\end{equation}

\begin{equation}
\lim_{\beta\to\infty} p^- = \lim_{\beta\to\infty} \frac{1}{1 + e^{-\beta [V(x_0 + 1) - V(x_0 - 1)]} + e^{-\beta [ V(x_0) - V(x_0 - 1) ]}}
= \frac{1}{1 + \infty + \infty} = 0
\end{equation}

Similar but opposite to the previous case, here all particles are pushed forward.

We can simplify the random walk problem using an energy scale argument as shown in the to limits above. We have seen that for systems with particles with high internal energies compared to the potential energy, they overrule the potential such that the particle does not have a preferred direction. Instead of computing the potential energies in such cases, we can just set the probabilities to one third for each option, and make the simulation faster by saving computational time.

Similarily, for potentials overriding the internal energies of the particles, we can set the probability to $100%$ in the direction of decreasing potential.

Rereading the tasks, we understand that we are asked to use the simplification described in exercise 1c, where we set the probability in the direction of decreased potential to &100%$. This would probably decrease the time it takes to run our program. However, we have decided to keep the numerical computation of probabilities, instead os assuming $100%$ probability for large $\beta k$. Especially when comparing our results from simulations to the analytical solution for the particle current, we would like to see how well they correspond without making further assumptions. This way, we can properly test whether the assumptions made in teh derivation of the analytical solution holds.

```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfc #complimentary error function
import ipywidgets as widgets #must be installed to display interactive plot in 4a!
from IPython.display import display
```

```python
#h = 1 and dt = 1 is set as default parameters in the random walk functions. integer values.

def V_i(x, alpha, Nx): #including alpha, Nx to be able to call any of the potential functions in later exercises
    """ 
        x: position along x-axis
        alpha: potential parameter for V1 function
        Nx: periodicity of V1 function

        Potential function for constant potential V i).
    """
    return 1 #k is included as beta_k in the probabilities function

def V_ii(x, alpha, Nx): #including alpha, Nx to be able to call any of the potential functions in later exercises
    """ 
        x: position along x-axis
        alpha: potential parameter for V1 function
        Nx: periodicity of V1 function

        Potential function for x-dependent potential V ii).
    """
    return -x #k is included as beta_k in the probabilities function

def logsumexp(exponents):
    """ Using LogSumExp trick to prevent overflow and underflow, used for computing probabilities
    
    Args:
        exponents (numpy array): lists of exponents we want to take LogSumExp of.

    Returns:
        float or numpy array : LogSumExp of input array
    """
    m = np.max(exponents)
    return m + np.log(np.sum(np.exp(exponents - m)))

def compute_probability(beta_k, x, V, alpha, Nx):
    """ Computes the probability for particle to move right or left in potential landscape
    
    Args:
        x (int or numpy-array): current position of particle
        V (function): the potential the particle lives in
        beta_k (int or float): scaling temperature and strength of potential
        alpha (float in range 0-1): determines position of sawtooth-peak in ratchet potential
        Nx (int): width of one sawtooth in ratchet potential

    Returns:
        (float, float): probability for moving right and left respectively
    """


    E_stay = -beta_k * V(x,alpha, Nx)
    E_plus = -beta_k * V(x+1,alpha, Nx)
    E_minus = -beta_k * V(x-1,alpha, Nx)

    z = logsumexp(np.array([E_stay,E_plus,E_minus]))

    p_plus  = np.exp(E_plus - z)
    p_minus = np.exp(E_minus - z)

    return p_plus, p_minus

def random_walk_step(Np, h, initial_positions, beta_k, V, alpha, Nx): 
    """ 
        Np: number of particles
        h: spatial step length
        initial_positions: a numpy array containing x-value for the position of each particle
        beta_k: beta * k, scalar number
        V: a potential function
        alpha: parameter of V1 function
        Nx: periodicity of V1 function

        Takes a random walk step (no particle interactions, no ratchet potential) for each particle, and updates array of positions in-place.
    """
    p = np.random.uniform(0,1,Np)
    for i in range(Np):
        p_plus, p_minus = compute_probability(beta_k, initial_positions[i], V, alpha, Nx)
        if(p[i]<=p_minus):
            initial_positions[i] -=h
        elif(p[i]>(1-p_plus)):
            initial_positions[i] +=h
    #updates in-place

def random_walk(Np, time_steps, beta_k, V, alpha, Nx, initial_positions, h=1, dt=1):
    """ 
    Args:
        Np: number of particles
        time_steps: number of time steps to complete ((number of cycles)*Tp)
        beta_k: beta * k, scalar number
        V: a potential function
        alpha: parameter of V1 function
        Nx: periodicity of V1 function
        initial_positions: numpy array of particles x-value at t=0
        h: spatial step length, default = 1
        dt: temporal step length, default = 1
    Returns:
        positions: array of positions after random walk is complete

    Completes random walk (no interactions, no ratchet potential) for the specified number of time_steps, and returns resulting positions array (copy). 
    """
    positions = initial_positions.copy()
    for i in range(0, time_steps, dt):
        random_walk_step(Np, h, positions, beta_k, V, alpha, Nx)
    return positions #each element in the array corresponds to one particle and holds the value of its position along the x-axis


def multiple_random_walks(time_steps, initial_positions, num_simulations, V, random_walk_func, beta_k, alpha, Nx, Np):
    """ Computes random walk for chosen number of simulations with time_steps timesteps each
    
    Args:
        time_steps (int): number of timesteps
        initial_positions(numpy array): array with x-value of each position at t=0
        num_simulations (int): number of random walk-simulations
        V (function): 1D potential landscape the particle lives in
        beta_k (int or float): scaling temperature and strength of potential
        alpha (float in range 0-1): determines position of sawtooth-peak in ratchet potential
        random_walk_func (function): lets us choose between different random-functions
        Nx (int): spatial periodicity of potential V1
        Np (int): number of particles

    Returns:
        numpy array : a concatenated array with end-distributions of all simulations of a random walk, for plotting histogram.
    """

    all_positions = [] #list to fill with positions after ranom walk

    for _ in range(num_simulations):
        all_positions.append(random_walk_func(Np, time_steps, beta_k, V, alpha, Nx, initial_positions))

    array_of_all_positions = np.concatenate(all_positions) #concatenating to a single array
    
    return array_of_all_positions


#COMPUTING VALUES:
V_i_positions_2a = multiple_random_walks(
    time_steps = 200, 
    initial_positions = np.arange(-50,51,1,dtype=int), 
    num_simulations = 100,
    V = V_i, 
    random_walk_func = random_walk,
    beta_k = 1, 
    alpha = None, #not in use here
    Nx = None, #not in use here
    Np = 101)

V_ii_positions_2a = multiple_random_walks(
    time_steps = 200, 
    initial_positions = np.arange(-50,51,1,dtype=int), 
    num_simulations = 100,
    V = V_ii, 
    random_walk_func = random_walk,
    beta_k = 1, 
    alpha = None, #not in use here
    Nx = None, #not in use here
    Np = 101)


#PLOTTING:
average_factor = np.full(100*101, 1/100) # 101 particles and 100 simulations
step_range = (-(200-50), (200+50)) # 200 time steps
bins = 200*2
fig1, (ax11, ax12) = plt.subplots(2,1)

ax11.hist(V_i_positions_2a, bins=bins, weights=average_factor, range=step_range)
ax11.grid()
ax11.set_title(f"Simulation of random walk with potential V(x)= k")
ax11.set_ylabel("Average number of particles")
ax11.set_xlabel("Position")

ax12.hist(V_ii_positions_2a, bins=bins, weights=average_factor, range=step_range)
ax12.grid()
ax12.set_title(f"Simulation of random walk with potential V(x)= -kx")
ax12.set_ylabel("Average number of particles")
ax12.set_xlabel("Position")

fig1.suptitle("Figure 1")
fig1.tight_layout()
plt.show()
```

We see from figure 1 that the particles in the uniform potential $V_i$ diffuse quite evenly from the initial distribusion, as expected when the probabilities are equal for all possible steps. The particles in potential $V_{ii}$ however, are all moved rightwards, towards lower potantial values.

```python
def random_walk_step_repulsion(h, current_positions, index, beta_k, V, alpha, Nx): 
    """ 
    h: spatial step length
    current_positions: numpy array with the x-value of each particle
    index: int, index of particle taking a step (in the array)
    beta_k: beta*k, scalar value
    V: a potential function
    alpha: parameter to V1 function
    Nx: spatial periodicity of V1

    Takes a step in the random walk simulation with hard repulsion, for one particle only. Returns new position value.
    """
    p = np.random.uniform(0,1)
    current_position = current_positions[index]
    p_plus, p_minus = compute_probability(beta_k, current_position, V, alpha, Nx)

    if(p<=p_minus):
        new_position = current_position - h
        if(check_if_available(current_positions, new_position)): #checks if new position is occupied
            return new_position #if available, returns the new position
        
    elif(p>(1-p_plus)):
        new_position = current_position + h
        if(check_if_available(current_positions, new_position)): #checks if new position is occupied
            return new_position
        
    return current_position #if no step is taken

def random_walk_repulsion(Np, time_steps, beta_k, V, alpha, Nx, initial_positions, h=1, dt=1): 
    """ 
        Np: number of particles (int)
        time_steps: number of steps to simulate
        beta_k: beta*k, scalar value
        V: a potential function
        initial_positions: numpy array with x-value of each particle at t=0
        alpha: parameter to V1 function
        Nx: spatial periodicity of V1
        h: spatial step length, default = 1
        dt: temporal step length, default = 1

        Complete the random walk simulation, with hard repulsion. Returns array of resulting positions after simulation.
    """
    positions = initial_positions.copy()
    particle_indices = np.arange(0, Np, dtype = int)

    for i in range(0, time_steps, dt):
        np.random.shuffle(particle_indices)
        for index in particle_indices:
            positions[index] = random_walk_step_repulsion(h, positions, index, beta_k, V, alpha, Nx) #taking random walk step and updating position
            
    return positions #each element in the array corresponds to one particle and holds the value of its position along the x-axis

def check_if_available(particle_positions, position_to_check):
    """ 
        particle_positions: numpy array of the x-value og each particles position
        position_to_check: scalar value (int) of a position along the x-axis which we want to check if is occupied   

        Checks how many particles are placed at the given position. (If repulsion: max 1)
        Returns boolean value: true if position is available, false if there is one (or more) particles occupying it. 
    """
    count = np.sum(particle_positions == position_to_check) #checking how many particles are at the given position
    return count == 0 #returns true if position is available

#COMPUTING VALUES:
V_i_positions_repulsion_2b = multiple_random_walks(
    time_steps = 200, 
    initial_positions = np.arange(-50,51,1,dtype=int), 
    num_simulations = 100,
    V = V_i, 
    random_walk_func = random_walk_repulsion,
    beta_k = 1, 
    alpha = None, #not in use here
    Nx = None, #not in use here
    Np = 101)

V_ii_positions_repulsion_2b = multiple_random_walks(
    time_steps = 200, 
    initial_positions = np.arange(-50,51,1,dtype=int), 
    num_simulations = 100,
    V = V_ii, 
    random_walk_func = random_walk_repulsion,
    beta_k = 1, 
    alpha = None, #not in use here
    Nx = None, #not in use here
    Np = 101)

#PLOTTING:
#average factor, bins and step range same as in 2a
fig2, (ax21, ax22) = plt.subplots(2,1)

ax21.hist(V_i_positions_repulsion_2b, bins=bins, weights=average_factor, range=step_range)
ax21.grid()
ax21.set_title(f"Simulation of random walk with repulsion and potential V(x)= k")
ax21.set_ylabel("Average umber of particles")
ax21.set_xlabel("Position")

ax22.hist(V_ii_positions_repulsion_2b, bins=bins, weights=average_factor, range=step_range)
ax22.grid()
ax22.set_title(f"Simulation of random walk with repulsion and potential V(x)= -kx")
ax22.set_ylabel("Average umber of particles")
ax22.set_xlabel("Position")

fig2.suptitle("Figure 2")
fig2.tight_layout()
plt.show()
```

Observations: 
Comparing the random walk simulations with repulsion (Figure 2) to the ones without (Figure 1), we see a more significant difference for the distributions living in a non-uniform potential ($V_{ii}$).

For the particles living in $V_i = k$, the average position on the x-axis is almost equal to the one in Figure 1, and so is the spread. The only difference is that we dont have spikes higher than 1 in the simulation with hard core repulsion, because the particles are not allowed to move to a position already occupied. The bars are a more evenly distributed here. 

For the particles living in $V_{ii} = -kx$, we have significantly different shapes of the distributions. The potential favours rightward movement. We see that both simulations in this potential $V_{ii}$, are shifted to the right compared to the ones in $V_i$. Due to repulsion and randomized order, the particles in this simulation must wait until all the particles to the right of them have moved. It generates a queue and accumulation in the back.
```python
def V1(x, alpha, Nx):
    """ 
        Sawtooth potential function.
        Args:
            x (int or numpy array): particle position(s)
            alpha (float in range (0,1)): parameter, decides peaks
            Nx (int): spatial periodicity, width of one sawtooth in ratchet potential 

        Returns:
            potential (scalar value, int or float) at given position x.
    """
    upper_lim = alpha*Nx
    lower_lim = -(1-alpha)*Nx

    x = upper_lim - (upper_lim - x)%Nx #open interval on lower limit, includes upper limit
    #V1 periodicity
    if (0<x<=upper_lim):
        return x/(alpha*Nx)#excluding k since k incorporated in probabilities
    elif (lower_lim<x<=0):
        return -x/((1-alpha)*Nx)


def V2(x, alpha, Nx): #taking in alpha,x,  Nx here as well to be able to call any of the potential functions in the probabilities function
    """ 
        x: x-value
        alpha: parameter og V1, decides peaks
        Nx: spatial periodicity

        Returns 1 for all x.
    """
    return 1 #constant value, set equal to 1 considering the magnitude is not important

def J(n_plus, n_minus, Np): #normalized potential current
    """ 
        n_plus: number of particles that took a step to the right during one time step
        n_minus: number of particles that took a step to the left during one time step
        Np: number of particles

        Returns the normalized potential current for time interval t(i) - t(i-1) (one time step).
    """

    return (n_plus-n_minus)/Np 

def random_walk_step_ratchet(Np, h, positions, beta_k, V, x_min, x_max, alpha, Nx): 
    """ 
        Np: number of particles
        h: spatial step length
        positions: numpy array of current positions (x-value of each particle)
        beta_k: beta*k, scalar value related to temperature (beta = 1/k_b*T) and potential strength (k)
        V: a potential function
        x_min: minimum x-value of x-axis
        x_max: maximum x-value of x-axis
        alpha: parameter of V1 function
        Nx: periodicity of V1 function

        Performs random walk (one time step) for all particles, and updates positions (in-place). Returns number of steps taken to the right and to the left (ints). 
    """
    p = np.random.uniform(0,1,Np)
    n_plus, n_minus = 0, 0
    for i in range(Np):
        p_plus, p_minus = compute_probability(beta_k, positions[i], V, alpha, Nx)

        if(p[i]<=p_minus):
            if(positions[i]==x_min):
                positions[i] = x_max #loops around if moving to the left from the leftmost point
            else:
                positions[i] -=h
            n_minus +=1 #counting number of particles taking a step to the left in this random walk step

        elif(p[i]>(1-p_plus)):
            if(positions[i]==x_max):
                positions[i] = x_min #loops around if moving right from rightmost point
            else:
                positions[i] +=h
            n_plus += 1

    return n_plus, n_minus


def random_walk_ratchet(x_axis, initial_positions, Np, time_steps, beta_k, V1, V2, Tp, Nx, alpha, h=1, dt=1): 
    """ 
        x_axis: numpy array of x-values along x-axis
        initial_positions: numpy array containing x-value of each particle at t= 0. (length = Np)
        Np: number of particles
        time_steps: number of random walk steps to be completed (one per time step)
        beta_k: beta*k, scalar value related to temperature (beta = 1/k_b*T) and potential strength (k)
        V1: one of two alternating potentials
        V2: second of two alternating potentials
        Tp: number of steps per half cycle
        Nx: periodicity of V1 function
        alpha: parameter of V1 function
        h: spatial step length, default = 1
        dt: temporal step length, default = 1

        Performs random walk for all particles for the given number of steps, and updates positions (copy). Returns list of average currents per cycle, but not the positions. 

    """
    x_min = np.min(x_axis)
    x_max = np.max(x_axis)

    positions = initial_positions.copy() #copying in order to preserve initial_positions array in case the same initial position-array is to be used multiple timess
    V = V2 #start in V2 at t = 0
    sum_particle_current = 0
    average_currents = []
    for t in range(0, time_steps, dt): 
        cycle_end = False
        if ((t+1)%(2*Tp) == 0):#if the next time step is an integer multiple of Tp, we are at the end of a cycle
            cycle_end = True
        elif (t%Tp == 0 and V == V2 and t!=0):#checking if at the end of a half cycle, if so the potential should be switched
            V = V1
        elif (t%Tp == 0 and V == V1 and t!=0):
            V = V2

        n_p, n_m = random_walk_step_ratchet(Np, h, positions, beta_k, V, x_min, x_max, alpha, Nx) #returuns number of particles that were moved right(np)/left(nm), positions are updaten in-place
        
        sum_particle_current += J(n_p, n_m, Np) #summing up particle currents

        if(cycle_end): #if at the end of a cycle, calculate and append average current
            J_av = sum_particle_current/(2*Tp)
            average_currents.append(J_av)
            sum_particle_current = 0 #ready for next cycle
            
    return average_currents #in this excercise we do not need to return the specific positions, only the particle currents

#COMPUTING VALUES:
#alpha = 0.8 and 0.1
average_currents_3a_alpha08 = random_walk_ratchet(
    x_axis = np.arange(0, 2*100, 1), #x-axis, two full sawtooth peaks
    initial_positions = np.repeat(np.arange(0, 2*100, 1), 6), #particles fill x-axis, 6 particles in each spot
    Np = 12*100, #12*Nx
    time_steps = 20*500, #20 cycles
    beta_k = 1000, 
    V1 = V1,
    V2 = V2, 
    Tp = 500, 
    Nx = 100, 
    alpha = 0.8)

average_currents_3a_alpha01 = random_walk_ratchet(
    x_axis = np.arange(0, 2*100, 1), 
    initial_positions = np.repeat(np.arange(0, 2*100, 1), 6), 
    Np = 12*100, 
    time_steps = 20*500, 
    beta_k = 1000, 
    V1 = V1, 
    V2 = V2, 
    Tp = 500, 
    Nx = 100, 
    alpha = 0.1)



#PRINTING AND PLOTTING:
print("For alpha = 0.8:")
for i in range(len(average_currents_3a_alpha08)):
    print(f"Cycle number {i+1}: average particle current = {average_currents_3a_alpha08[i]}")
cycles_3a_alpha08 = [i for i in range (1,len(average_currents_3a_alpha08)+1)] #x-axis
plt.plot(cycles_3a_alpha08, average_currents_3a_alpha08, label = "alpha = 0.8")

print("For alpha = 0.1:")
for i in range(len(average_currents_3a_alpha01)):
    print(f"Cycle number {i+1}: average particle current = {average_currents_3a_alpha01[i]}")
cycles_3a_alpha01 = [i for i in range (1,len(average_currents_3a_alpha01)+1)] #x-axis
plt.plot(cycles_3a_alpha01, average_currents_3a_alpha01, label = "alpha = 0.1")

plt.title("Figure 4 \nAverage particle current per cycle for two different values of alpha")
plt.grid()
plt.legend()
plt.xlabel("Cycle number")
plt.ylabel("Average particle current")
plt.show()

#Plotting the sawtooth potential for different alpha for better understanding:

#computing potential values
x_axis_3 = np.arange(0, 2*100, 1)

V1_vals_alpha08 = []
V1_vals_alpha01 = []
V1_vals_alpha05 = []
for x in x_axis_3: #x-axis
    V1_vals_alpha08.append(V1(x, 0.8, 100)) #alpha = 0.8, Nx = 100
    V1_vals_alpha01.append(V1(x, 0.1, 100)) #alpha = 0.1, Nx = 100
    V1_vals_alpha05.append(V1(x, 0.5, 100)) #alpha = 0.5, Nx = 100

plt.figure()
plt.title("Figure 5 \nV1(x, alpha)")
plt.plot(x_axis_3, V1_vals_alpha01, label="alpha = 0.1")
plt.plot(x_axis_3, V1_vals_alpha05, label="alpha = 0.5")
plt.plot(x_axis_3, V1_vals_alpha08, label="alpha = 0.8")
plt.xlabel("x")
plt.ylabel("V1(x)")
plt.legend()
plt.grid()
plt.show()
```

The plot in Figure 6 shows a clear tendency of increased average current (again, in absolute value), as the number of timesteps in each cycle increases. When reaching a certain point, the increase flattens. For our plots, this happens for $T_p$-values around $300-400$. With $p^+=p^0=p^-=\frac{1}/{3}$, we can compute the expectation value and the variance of one particles position after 300 steps. 

$X$ is the event of one step, $Y$ is the event of 300 independent steps.

Expectation value for one step, $E(X) = 1\cdot\frac{1}{3} + 0\cdot\frac{1}{3} + -1\cdot\frac{1}{3} = 0$

This gives an expectation value for 300 steps equal to $E(Y) = \sum_1^{300}E(X) = 300\cdot E(X) = 0$

For computing variance, we first compute $E(X^2) = 1^2\cdot\frac{1}{3} + 0\cdot\frac{1}{3} + (-1)^2\cdot\frac{1}{3} = 2/3$

This gives $Var(X) = E(X^2)-(E(X))^2 = \frac{2}{3}$

Since we have a sum of independent steps, we obtain $Var(Y) = \sum_1^{300} Var(X) = 300 \cdot \frac{2}{3} = 200$

This equals a standard deviation $SD(Y) = \sqrt{200} \approx 14$

The standard deviation tells us within what distance from the expectation value, most of the particles will end up. Since the distance to the nearest maxima from a potential minima is equal to $20$ with $\alpha = 0.8$ and $Nx = 100$, it makes sense that it is for values larger than this, the curve starts to stabilize. As the standard deviation approaches 20, significantly more particles will diffuse and reach long enough to slide over to the neighbouring minima when the potential switches from uniform to ratchet. This conclution is partially based on the assumption that we have sufficiently many observations to assume that our particles follows a normal distribution, according to the central limit theorem.

During the diffusion, for small enough values og $T_p$, the particles are not likely to move past the maxima and will therefore stay within the same potential valley. For the particles that diffuse to some point on the slopes of the valley, the probability of them falling back down towards the same minima, when the V1 potential is switched on, is near $100%$. However, for a particle placed *at* the minima after diffusion, there will be a slight probability of it moving right, where the slope is less steep. Therefore, even for large $\beta k$, for small values of $T_p$, we expect to observe a very weak net current to the right. 

This matches our observations for the average cycle current for $T_p = 21, 41$. For $T_p = 1$ we observe that the average current fluctuates a lot between positive and negative values, which can seem unphysical. This can be a result of the small time period, $T_p = 1$. Since we only simulate one cycle, a maximum of two steps is taken by each particle. There is no period of time where the particles will stay still in a favourable configuration, because the potential switches every step.

```python
def analytical_J(Nx, Tp, alpha):
    """ 
        Nx: spatial periodicity of V1 in ratchet potential
        Tp: number of time steps per half cycle
        alpha: parameter of V1-function, decides sawtooth peaks

        Returns analytical values of J_avg.
    """
    return Nx/(4*Tp) * (erfc(alpha*Nx / 2*np.sqrt(3/Tp)) - erfc((1-alpha)*Nx / 2*np.sqrt(3/Tp)))


def alpha_dependent_particle_current(x_axis, initial_positions, Np, time_steps, beta_k, V1, V2, Tp, Nx, alpha_values):
    """ Random walk simulation for one cycle for different values of alpha.
    
    Args:
        x_axis: numpy array of x-values
        initial_positions (numpy array): start distribution of the particles. Each index represents a particle, and the value its position.
        Np (int): number of particles
        time_steps: total number of time steps
        Tp: number of time steps per half cycle
        beta_k (int or float): scaling temperature and strength of potential
        alpha_values (numpy array or list): containing alpha values for different cycles. Determines position of sawtooth-peak in ratchet potential
        V1: potential function V1
        V2: potential function V2
        
    Returns:
        (list) : particle current for simulations of a set of alpha values
    """
    particle_current_different_alphas = []

    for alpha in alpha_values: #only one cycle -> one element in return value from random_walk_ratchet
        [particle_current] = random_walk_ratchet(x_axis, initial_positions, Np, time_steps, beta_k, V1, V2, Tp, Nx, alpha)
        particle_current_different_alphas.append(particle_current)
    
    return particle_current_different_alphas



#COMPUTATION:
initial_positions_3c = np.concatenate((V1min1*np.ones(12*100//2), V1min2*np.ones(12*100//2))) #starting with half of the particles in each potential minima
alpha_vals_3c = np.linspace(0.01,1,50)

J_av_3c = alpha_dependent_particle_current(
    x_axis = x_axis_3, 
    initial_positions = initial_positions_3c,
    Np = 12*100,
    time_steps = 2*500,
    beta_k = 1000,
    V1 = V1,
    V2 = V2, 
    Tp = 500, 
    Nx = 100, 
    alpha_values = alpha_vals_3c)

#PLOTTING:
plt.plot(alpha_vals_3c, J_av_3c, label = "Simulation", color = "darkblue")
plt.plot(alpha_vals_3c, analytical_J(100, 500, alpha_vals_3c), label="Analytical", color = "lightsteelblue")
plt.title("Figure 7 \nAverage particle current as a function of alpha")
plt.ylabel(r"$J_{\text{avg}}$")
plt.xlabel(r"$\alpha$")
plt.legend()
plt.grid()
plt.show()
```

```python
def step_left_ratchet_repulsion(position, x_min, x_max, b, positions_array, h):
    """ 
        Args:
            position: int, current position (x-value)
            x_min: left limit on x-axis
            x_max: righgt limit on x-axis
            b: particles physical extension (scalar value)
            positions_array: array of each particles current position/x-value
            h: spatial step length

        Returns:
            new_position: new position if step is taken
            available: boolean value, true if step can be taken (no particle will collide)
        
        Function to take a step to the left in random_walk_step_ratchet_repulsion.
    """
    if(position==x_min):
        new_position = x_max #loops around if moving to the left from the leftmost point
    else:
        new_position = position - h
        
    position_to_check = new_position - np.floor(b) #checking whether there is another particle that can cause collision
    if (position_to_check < x_min): #looping boundary conditions
        position_to_check = x_max - np.abs(position_to_check) + h 

    available = check_if_available(positions_array, position_to_check) #returns true if no particle will collide when step is taken
    return new_position, available


def step_right_ratchet_repulsion(position, x_min, x_max, b, positions_array, h):
    """ 
        Args:
            position: int, current position (x-value)
            x_min: left limit on x-axis
            x_max: righgt limit on x-axis
            b: particles physical extension (scalar value)
            positions_array: array of each particles current position/x-value
            h: spatial step length

        Returns:
            new_position: new position if step is taken
            available: boolean value, true if step can be taken (no particle will collide)
        
        Function to take a step to the right in random_walk_step_ratchet_repulsion.
    """
    if(position==x_max):
        new_position = x_min #loops around x-axis if moving right from rightmost point
    else:
        new_position = position + h

    position_to_check = new_position + np.floor(b) #position that needs to be free in order to accept the step
    if (position_to_check > x_max): #looping boundary conditions (the repulsion from a particles physical extension loops around the x-axis)
        position_to_check = position_to_check - x_max - h
    
    available = check_if_available(positions_array, position_to_check) #returns true if no particle will collide when step is taken
    return new_position, available


def random_walk_step_ratchet_repulsion(Np, h, positions, beta_k, V, x_min, x_max, b, alpha, Nx):
    """ 
        Np: number of particles
        h: spatial step length
        positions: numpy array of current positions
        beta_k: beta*k, scalar value related to temperature and potential
        V: a potential function
        x_min: lowest x-value on x-axis (x-axis loops around these values)
        x_max: highest x-value on x-axis
        b: scalar number, physical extension (length) of particle
        alpha: parameter for V1 function
        Nx: spatial periodicity of V1 function

        Performs one step for all particles in the random walk simulation, particle by particle, in random order, and updates changed positions in place. Also returns number of steps taken to the right and to the left.
    """
    p = np.random.uniform(0,1,Np)
    n_plus, n_minus = 0,0
    particle_indices = np.arange(0,Np, dtype = int)
    np.random.shuffle(particle_indices) #shuffling indices so that the particles will attempt to move one by one in randomized order
    
    for i in particle_indices:
        position = positions[i]
        p_plus, p_minus = compute_probability(beta_k, position, V, alpha, Nx)

        if(p[i]<=p_minus):
            new_position, is_available = step_left_ratchet_repulsion(position, x_min, x_max, b, positions, h)
            if(is_available): #if no particle to collide with, the step is executed
                positions[i] = new_position #updating position in-place
                n_minus +=1 #counting number of particles taking a step to the left in this random walk step

        elif(p[i]>(1-p_plus)):
            new_position, is_available = step_right_ratchet_repulsion(position, x_min, x_max, b, positions, h)
            if(is_available):
                positions[i] = new_position
                n_plus += 1

    return n_plus, n_minus #returning the number of particles that have taken a step right and left


def random_walk_ratchet_repulsion(x_axis, initial_positions, Np, time_steps, beta_k, V1, V2, Tp, Nx, b, alpha, h=1, dt=1):
    """ 
        x_axis: numpy array containing all x-values (preferrably ints, in order to be able to check if equal to another position)
        initial_positions: numpy array containing x-value of all particles at t=0
        Np: number of particles
        time_steps: number of time steps to run the random walk for
        h: spatial step length
        dt: temporal step length
        beta_k: beta*k, scalar value related to temperature and potential
        V1: first of two alternating potentials (function)
        V2: second of two alternating potentials (function)
        Tp: time steps per half cycle
        Nx: spatial periodicity of V1 function
        b: scalar number, physical extension (length) of particle
        alpha: parameter for V1 function

        Performs random walk for given number of steps, for all particles, particle by particle, in random order, and updates changed positions. Returns new array of positions after random walk is complete, as well as a list of average currents per cycle, and a 2D-list containing (copies of) arrays of positions at each time step.
    """
    positions = initial_positions.copy()
    x_min = np.min(x_axis)
    x_max = np.max(x_axis)
    positions_over_time = []
    V = V2 #start in V2 at t=0
    sum_particle_current = 0
    average_currents = []
    for t in range(0,time_steps,dt):
        cycle_end = False
        if ((t+1)%(2*Tp) == 0): #if the next time step is an integer multiple of 2*Tp, we are at the last time step of a cycle
            cycle_end = True
        elif (t%Tp == 0 and V == V2 and t!=0): #potential switches every half cycle. If t is an integer multiple of Tp, we are at the end of a half cycle and need to switch potential
            V = V1
        elif (t%Tp == 0 and V == V1 and t!=0):
            V = V2
        n_p, n_m = random_walk_step_ratchet_repulsion(Np, h, positions, beta_k, V, x_min, x_max, b, alpha, Nx) #returns number of particles that were moved right(np)/left(nm). positions are updated in-place
        sum_particle_current += J(n_p, n_m, Np) #summing up particle currents
        positions_over_time.append(positions.copy())

        if(cycle_end): #if at the end of a cycle, calculate average current
            J_av = sum_particle_current/(2*Tp)
            average_currents.append(J_av)
            sum_particle_current = 0 #ready for nexyt cycle
    
    return positions, average_currents, positions_over_time


#COMPUTATIONS
x_axis_4a = np.arange(0, 4*20, 1) #x-axis, Nx = 20, 4 full sawtooth peaks -> 4*20 points
initial_positions_4a = np.linspace(4, 4*20-4, 10, dtype = int) #even distribution, few enough particles to avoid overlap. b=4. Np = 10 values.

_, _, positions_over_time_4a = random_walk_ratchet_repulsion(
    x_axis = x_axis_4a, 
    initial_positions = initial_positions_4a, 
    Np = 10, 
    time_steps = 5*2*40, 
    beta_k = 1000, 
    V1 = V1, 
    V2 = V2, 
    Tp = 40, 
    Nx = 20, 
    b = 4, 
    alpha = 0.2)

#PLOTTING
positions_over_time_4a = np.array(positions_over_time_4a)
plt.figure()
#plt.grid()
for x_val in range(0,400,40):
    plt.axvline(x_val, linestyle = "--", color = "darkgray") #plotting where potentials shift (at the half cycles)

#plotting position of each particle at each time step
for i in range(10): #Np = 10, plot for each particle
    plt.scatter([j for j in range(len(positions_over_time_4a))],  positions_over_time_4a[:,i]) #slicing to find each particle's position

plt.title("Figure 11 \nPosition of each particle over time")
plt.xlabel("Time steps")
plt.ylabel("Particle position (x-axis)")
plt.show()
```

```python

def get_density(b, Np, Ns, Nx):
    """ Finds particle density based on number of particles and start distribution.

    Args:
        b (int or float): the width/size of one particle.
        Np (int): number of particles
        Ns (int): Numper of repeating periods in ratchet potential
        Nx (int): width of one sawtooth in ratchet potential
        
    Returns:
        (float) : particle density
    """
    return b*Np/(Ns*Nx)

def get_Np_max(b, Ns, Nx, h=1):
    """ Finds maximal numbers of particles that can fit into start distribution

    Args:
        b (int or float): the width/size of one particle.
        Ns (int): Numper of repeating periods in ratchet potential
        Nx (int): width of one sawtooth in ratchet potential
        h (int): spatial step length, default = 1
        
    Returns:
        (int) : maximal number of particles
    """
    steps_until_next = np.floor(b)+h
    return Ns*Nx//steps_until_next


def create_initial_positions_repulsion(b, Np, x_axis, h=1):
    """ 
        Args:
            h: spatial step length, default = 1
            b: (b) physical extension (length) of particle
            Np: number of particles in system
            x_axis: numpy array containing all x-values (preferrably ints, in order to be able to check if equal to another position)
        
        Returns:
            initial_positions (numpy array): initial positions for a given Np for particles with repulsion and extension
            
        Creates and returns an initial position array for particles with hard repulsion and a physical extension where they are placed as far away from each other as possible while still upholding the repulsion "rules".
    """
    x_min = np.min(x_axis)
    initial_positions = np.zeros(Np)
    distance_between = (len(x_axis)-1)//Np #maximum distance between particles without creating overlap
    distance_occupied = h + np.floor(b)
    first_particle = x_min + (distance_occupied)//2 #setting first particle with left side at x_min, placing particles as spread out as possible without causing overlap
    initial_positions[0] = int(first_particle)
    x_curr = first_particle
    for i in range(1,Np):
        position = x_curr + distance_between
        x_curr = position
        initial_positions[i] = int(position)
    return initial_positions


def get_initial_positions_densities(Ns, Nx, b, x_axis):
    """ Returns start distribution for different densities so particles dont overlap, as well as a list of densities.

    Args:
        Ns (int): Numper of repeating periods in ratchet potential
        Nx (int): width of one sawtooth in ratchet potential
        b (int or float): the width/size of one particle.
        x_axis: numpy array containing all x-values (ints)

        
    Returns:
        (list of numpy arrays) : startdistributions and corresponding densities
        (list): list of densities
    """
    Np_max = get_Np_max(b, Ns, Nx)
    densities_list = []
    initial_positions_list = []
    for Np in range(1, Np_max+1,4): #step=4 so we dont evaluate too many densities.
        #current_start_dist = np.linspace(0, Ns*Nx - (np.floor(b)+1), Np, dtype=int)
        initial_positions = create_initial_positions_repulsion(b, Np, x_axis)
        initial_positions_list.append(initial_positions)

        rho = get_density(b, Np, Ns, Nx)
        densities_list.append(rho)
    
    return initial_positions_list, densities_list


#COMPUTATIONS
x_axis_4b = np.arange(1, 100*10, 1, dtype=int) #creating x-axis, Ns*Nx = 10*100 points

initial_positions_list_4b, densities_4b = get_initial_positions_densities(
    Ns = 10,
    Nx = 100,
    b = 20,
    x_axis = x_axis_4b)


cycle_averaged_currents_4b = []
for initial_positions_arr in initial_positions_list_4b:
    Np_value = len(initial_positions_arr) #count number of particles
    _, average_currents_4b, _ = random_walk_ratchet_repulsion(
        x_axis = x_axis_4b, 
        initial_positions = initial_positions_arr, 
        Np = Np_value, 
        time_steps = 100*2*300,#(100 cycles) * (300*2 steps per cycle) 
        beta_k = 1000, 
        V1 = V1, 
        V2 = V2, 
        Tp = 300, 
        Nx = 100, 
        b = 20, 
        alpha = 0.2)
    cycle_averaged_currents_4b.append(np.mean(average_currents_4b)) #adding the average over all 100 cycles for one value of rho to a list

#PLOTTING
plt.figure()
plt.plot(densities_4b, cycle_averaged_currents_4b)
plt.grid()
plt.title("Figure 13 \nCycle averaged currents for Tp = 300 as function of density")
plt.xlabel(r"$\rho$")
plt.ylabel(r"$J_{\text{avg}}$")
plt.show()
```

Figure 13 shows the cycle averaged particle currents for various density values. The particle current is positive because $\alpha = 0.2$, as explained in exercise 3a.
One would think that higher density leads to lower average current, since each particle then has less space to move around. This is exactly the trend we see for densities higher than 0.4. 

The increase from lower to intermediate densities seems unphysical at first, but taking a closer look, we belive we have found a possible explanation. For the lowest values of $\rho$, each particle has a lot of cleared space around them, which makes them free to move in the direction they want. They easily fall into a potential minima, but then often get stuck there. The distance they need to diffuse in order to reach over to the neighbouring minima, is long and therefore statisticlly less likely for the particle to reach (ref. exercise 3b). For intermediate values of $\rho$, the particles can still move around, but they will interact and begin forming a "queue". Some particles are stopped higher up in the valley because the lower placements are occupied. For these particles, fewer random steps are required for them to fall into the neighbouring valley when the potential switches. This is the kind of movement that causes a net current. For even larger density values, there is not enough room for the particles to move, and the current falls again.
